```python
import splusdata
import getpass
import pandas as pd
from astropy.table import Table, vstack
import numpy as np
from pathlib import Path
  
ROOT = Path("../WD-GaiaDr3") 
```

```python
# Connecting with SPLUS database
username = input(prompt="Login: ")
password = getpass.getpass("Password: ")
conn = splusdata.connect(username, password)
```

Reading the csv table

```python
df = pd.read_csv(ROOT / "WD-gaiadr3-p95-cor03.csv")
```

```python
colums = ["RA_ICRS", "DE_ICRS"]
df_ = df[colums]
df_
```

```python
Query = f"""SELECT ref.ID, ref.ra, ref.dec FROM "idr3"."all_idr3" 
            AS ref JOIN TAP_UPLOAD.upload AS upl 
            ON 1=CONTAINS( POINT('ICRS', ref.ra, ref.dec), 
            CIRCLE('ICRS', upl.RA_ICRS, upl.DE_ICRS, 0.000277777777778))"""
```

```python
n = int(len(df_) / 100.) + 1
print(n)
```

```python
df_f = [] # list
j = 0 # counter
d = {} # empty
for i in range(n):
    j += 1  
    df_f.append(df_.iloc[100*i:100*j])
```

```python
merged_table_list = []
for a in range(n):
    results = conn.query(Query, df_f[a])
    merged_table_list.append(results)
```

```python
merged_table_list
```

```python
# Merging all result astropy tables 
from astropy.table import vstack
merged_table = vstack(merged_table_list)
```

```python
len(merged_table)
```

```python
merged_table.write(ROOT / "WD-gaiadr3-p95-SPLUS-cor03.ecsv", format="ascii.ecsv", overwrite=True)
df_table = merged_table.to_pandas()
df_table.to_csv(ROOT / "WD-gaiadr3-p95-SPLUS-cor03.csv", index = False)
```
